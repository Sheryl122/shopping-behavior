```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
shopping_data = pd.read_csv('data/processed/shopping_cleaned.csv')
```

```python
shopping_data.head()
shopping_data.info()
shopping_data.describe().round(2)
```

```python
shopping_data.groupby('Payment Method').size().reset_index(name='Count').sort_values(by='Count', ascending=False)
```

## Which gender spends more in each season?


```python
shopping_data.pivot_table(values='Purchase Amount (USD)', index='Gender', columns='Season', aggfunc='mean').round(2)
```

## Which age group spends the most in each location?

```python
shopping_data.groupby('Location')['Age'].mean().astype(int)
```
