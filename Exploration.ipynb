# Exploration Questions

## Which payment methods yields the highest amount of purchases overall?
Based on the table below Credit Cards are most used for purchases. 

```python
shopping_data.groupby('Payment Method').size().reset_index(name='Count').sort_values(by='Count', ascending=False)
```

## Which gender spends more in each season?
During the Fall males spend more, during the Spring and winter non-binary people spend the most. For winter females spend the most. This information is important to understand which items will most likely sell the most during each season.


```python
shopping_data.pivot_table(values='Purchase Amount (USD)', index='Gender', columns='Season', aggfunc='mean').round(2)
```

```python
gender_x_season = shopping_data.pivot_table(values='Purchase Amount (USD)', index='Gender', columns='Season', aggfunc='mean').round(2)
gender_x_season.idxmax()
```

## Which age group spends the most in each location?
The chart below displays the average age of spenders in each state. This information allows FlashFash to cater to each age group in each state with intention.

```python
shopping_data.groupby('Location')['Age'].mean().astype(int).reset_index()
```

## Which Shipping Method is most used?
Based on these stats Standard shipping is the most used shipping method. Consumers tend to abandon their online cart once they discover that there is a shipping fee, so this information can help FlashFash come up with a balance between item cost and shipping cost so that the shipping can appear free to the consumer and they are more likely to complete their purchase.

```python
shopping_data.groupby('Shipping Type')['Item Purchased'].size().reset_index().sort_values(by='Item Purchased', ascending=False)
```
